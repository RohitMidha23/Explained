```python
players_by_weight = df.groupby('Weight').size()
#use .size() to get aggregated value of the particular column only or .count() for every column. 
#Change .size() to .count() and run to see difference
print(players_by_weight)
```

```python
xvals = [i for i in range(140,330) if i%20==0]
#range should be min value of x-axis label to max value +10 of x-label
plot_by_weight = players_by_weight.plot(title='Number of Players by Weight',xticks = xvals)
plot_by_weight.set_ylabel('Number of Players')
plot_by_weight.set_xlabel('Weight')
plot_by_weight
```

```python
players_by_age = df.groupby('Age').size()
print(players_by_age)
```

```python
plot_by_age = players_by_age.plot(title='Players by Age',xticks = range(18,42),rot=2)
plot_by_age.set_ylabel('Number of Players')
plot_by_age.set_xlabel('Age')
plot_by_age
# This shows that 
```
